```python
#taking 49 data
df=df[:49]
```

```python
for index, values in df.iterrows():
    print(values)
```

```python
for index,values in df.iterrows():
    name=values['NAME']
    dateofbirth=values['DATE OF BIRTH']
    gender=values['GENDER']
    bloodgroup=values['BLOOD GROUP']
    nationality=values['NATIONALITY']
    adhaarcardno=values['ADHAAR CARD NO.']
    permanentaddress =values['PERMANENT ADDRESS']
    district=values['DISTRICT']
    state=values['STATE']
    pincode=values['PINCODE']
    uploadcovid19certificate =values['UPLOAD COVID19 CERTIFICATE']
    
    data=f"""
        NAME: {name}
        DATE_OF_BIRTH: {dateofbirth}
        GENDER: {gender}
        BLOOD_GROUP: {bloodgroup}
        NATIONALITY: { nationality}
        ADHAAR_CARD_NO: {adhaarcardno}
        PERMANENT_ADDRESS: {permanentaddress}
        DISTRICT: {district}
        STATE: { state}
        PINCODE: {pincode}
        UPLOAD_COVID19_CERTIFICATE: { uploadcovid19certificate}
        """
    #creating qr code on the present data
    img=pyqrcode.create(data)
    
    save_file="+name+".svg"
    img.svg(save_file,scale=5)
```
